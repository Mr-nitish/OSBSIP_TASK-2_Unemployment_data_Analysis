```python
#importing useful Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import plotly.express as px
```

Loading and Reading Dataset

```python
data = pd.read_csv("Unemployment_in_India.csv")
```

Removing Null values

```python
data=data.dropna()
```

Changing Attribute Name and Plotting Correlation

```python
data.columns=['Regin','Date','Frequency','Estimate_rate_%','Estimate_Employed','Estimate_participation_rate_%','Area']
```

```python

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
sb.heatmap(data.corr())
plt.show
```

```python
plt.figure(figsize=(16,12))
plt.title("Indian Unemployment")
sb.histplot(x="Estimate_Employed", hue='Regin' , data=data)

plt.show()
```

```python
unemplyoment= data[['Regin','Date','Estimate_rate_%','Area','Estimate_Employed']]

figure = px.sunburst(unemplyoment , path=['Area','Regin'],
                     values ='Estimate_rate_%',
                     width=700,height=700, color_continuous_scale='RdY1Gn',
                     title= 'Unemployment Rate In India') 
figure.show()
```
